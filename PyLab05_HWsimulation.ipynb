```python
####################################################################
############## DO NOT CHANGE THE CODE IN THIS CELL #################
####################################################################

import numpy as np
import pandas as pd
from IPython.display import display

# LOADING THE DATASETS AS DATAFRAMES
myfifa_df = pd.read_csv('fifa19datastats.csv', index_col='Index')
skill_types_df = pd.read_csv('skilltypes.csv')
skill_cats_df = pd.read_csv('skillcategories.csv')

features = myfifa_df.columns.to_list()[4:]
features_by_type = skill_types_df.groupby('type')['skill'].apply(list).to_dict()
features_by_category = skill_cats_df.groupby('category')['skill'].apply(list).to_dict()
labels = myfifa_df.columns.to_list()[1:4]

print(f'*** THESE ARE THE {len(features)} FEATURES (SEE VARIABLE features):')
for ff in features:
    print(f'{ff}')
print('*************************************')
print('')
print('*** SELECTED FEATURES BY TYPES (SEE VARIABLE features_by_type):')
for tt in features_by_type.keys():
    print(f'{tt}: {features_by_type[tt]}')
    print('')
print('*************************************')
print('')
print('*** SELECTED FEATURES BY CATEGORY (SEE VARIABLE features_by_category):')
for tt in features_by_type.keys():
    print(f'{tt}: {features_by_type[tt]}')
    print('')
print('*************************************')
print('')
print('*** THESE ARE THE LABELS (SEE VARIABLE labels):')
for ll in labels:
    print(f'{ll}')
print('*************************************')

def which_rows(df, frac, *StudentIDs):
    random_seed = min(StudentIDs)
    df_ = df.sample(frac=frac, random_state=random_seed)
    return df_

myfifa = which_rows(myfifa_df, 0.75, *StudentIDs)
myfifa = myfifa.loc[:, features + labels]

myfifa_ft = myfifa.loc[:, features]
myfifa_lb = myfifa.loc[:, labels]

print('')
print('*** THIS IS YOUR PERSONAL DATASET (features AND labels TOGETHER, SEE VARIABLE myfifa)')
display(myfifa)
print('')
print('*** THIS IS YOUR PERSONAL DATASET (features, SEE VARIABLE myfifa_ft)')
display(myfifa_ft)
print('')
print('*** THIS IS YOUR PERSONAL DATASET (labels, SEE VARIABLE myfifa_lb)')
display(myfifa_lb)

random_seed = min(StudentIDs)
np.random.seed(random_seed)
```
